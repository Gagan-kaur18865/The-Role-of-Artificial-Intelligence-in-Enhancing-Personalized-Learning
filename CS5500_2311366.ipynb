```python
import pandas as pd
import scipy.stats as stats

# Step 1: Correlating Tool Usage and Outcomes
# Perform Chi-Square Test between usage of adaptive tools and personalized learning experience
adaptive_tools_keywords = ['adaptive learning', 'AI-driven assessments']
updated_data['Adaptive or Assessment Tools Used'] = updated_data['Which AI-powered tools have you used in your online learning/teaching experience? (Select all that apply)'].str.contains('|'.join(adaptive_tools_keywords), case=False, na=False)

# Create a crosstab for Adaptive Tools Used and Personalized Learning Experience
crosstab_personalized = pd.crosstab(updated_data['Adaptive or Assessment Tools Used'], updated_data['Do you feel that AI tools provide personalised learning experiences?'])

# Perform Chi-Square test
chi2_personalized, p_personalized, dof_personalized, ex_personalized = stats.chi2_contingency(crosstab_personalized)

# Print Chi-Square test result for Personalized Learning Experience
print(f"Chi-Square Test for Personalized Learning Experience and Adaptive Tools Usage:")
print(f"Chi-Square value: {chi2_personalized}")
print(f"P-value: {p_personalized}")
print(f"Degrees of Freedom: {dof_personalized}\n")

# Create a crosstab for Adaptive Tools Used and Engagement Impact
crosstab_engagement = pd.crosstab(updated_data['Adaptive or Assessment Tools Used'], updated_data['How has the use of AI tools impacted your engagement with the course material?'])

# Perform Chi-Square test for engagement
chi2_engagement, p_engagement, dof_engagement, ex_engagement = stats.chi2_contingency(crosstab_engagement)

# Print Chi-Square test result for Engagement Impact
print(f"Chi-Square Test for Engagement Impact and Adaptive Tools Usage:")
print(f"Chi-Square value: {chi2_engagement}")
print(f"P-value: {p_engagement}")
print(f"Degrees of Freedom: {dof_engagement}\n")

# Step 2: Further Refinement - Break down by demographics (e.g., Role, Age Group)

# Group by Role
for role, role_group in updated_data.groupby('What is your role in the educational institution?'):
    # Crosstab and Chi-Square Test for Personalized Learning Experience within each role
    crosstab_role_personalized = pd.crosstab(role_group['Adaptive or Assessment Tools Used'], role_group['Do you feel that AI tools provide personalised learning experiences?'])

    # Perform Chi-Square test
    chi2_role_personalized, p_role_personalized, dof_role_personalized, ex_role_personalized = stats.chi2_contingency(crosstab_role_personalized)

    # Print Chi-Square test result for Personalized Learning Experience within the role
    print(f"Chi-Square Test for Personalized Learning Experience and Adaptive Tools Usage (Role: {role}):")
    print(f"Chi-Square value: {chi2_role_personalized}")
    print(f"P-value: {p_role_personalized}")
    print(f"Degrees of Freedom: {dof_role_personalized}\n")

# Group by Age Group
for age_group, age_group_data in updated_data.groupby('What is your age group?'):
    # Crosstab and Chi-Square Test for Engagement Impact within each age group
    crosstab_age_engagement = pd.crosstab(age_group_data['Adaptive or Assessment Tools Used'], age_group_data['How has the use of AI tools impacted your engagement with the course material?'])

    # Perform Chi-Square test
    chi2_age_engagement, p_age_engagement, dof_age_engagement, ex_age_engagement = stats.chi2_contingency(crosstab_age_engagement)

    # Print Chi-Square test result for Engagement Impact within the age group
    print(f"Chi-Square Test for Engagement Impact and Adaptive Tools Usage (Age Group: {age_group}):")
    print(f"Chi-Square value: {chi2_age_engagement}")
    print(f"P-value: {p_age_engagement}")
    print(f"Degrees of Freedom: {dof_age_engagement}\n")
```
